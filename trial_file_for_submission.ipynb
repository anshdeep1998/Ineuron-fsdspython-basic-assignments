```python
class csv_bulk_upload:
    def __init__(self,filename):
        logging.info("initializing the file in constructor")
        self.filename = filename
        
    def give_lists(self):
        logging.info("Trying to read file")
        lo=[]
        try:
                with open(self.filename,'r') as f:
                    carbon_data = csv.reader(f,delimiter = '\n')
                    for i in enumerate(carbon_data):
                        for list_ in i[1]:
                            lo.append(i[1])
                return lo
        
        except Exception as e:
            logging.error("Error occurred in opening this file")
            logging.exception("The error occurred is: ",e)
        
        
    def give_column_names(self,list_of_rowlists):
        logging.info("fetching the list of names of all columns")
        
        try:
            column_names = list_of_rowlists[0][0].split(",")
            return column_names
        except Exception as e:
            logging.error("Error occurred in fetching list of columns")
            logging.exception("The error occurred is: ",e)
            
    
    def give_split_lists_of_row_values(self,list_of_rowlists):
        logging.info("All row values are split into lists")
        
        try:
            list_of_rowlists.remove(list_of_rowlists[0])
            l_new=[]
            for i in list_of_rowlists:
                split_rows = i[0].split(",")
                l_new.append(split_rows)
            return l_new
        
        except Exception as e:
            logging.error("Exception occurred in fetching split lists values")
            logging.exception("The error occurred is: ", e)
            
            
    def give_json_docs(self,split_lists_of_row_values,column_list):
        logging.info("All rows of csv will be converted to json docs")
        try:
            list_of_dicts=[]
            for split_lists in split_lists_of_row_values:
                dict_new={column_list[i]: split_lists[i] for i in range(len(column_list))}
                list_of_dicts.append(dict_new)
        
            return list_of_dicts
        
        except Exception as e:
            logging.error("Exception occurred in converting to jsons")
            logging.exception("The error occurred is: ", e)
            
    def bulk_insert(self,collection_name,list_of_json_docs):
        logging.info("Here comes insertion of all json docs into database of mongoDB")
        try:
            collection_name.insert_many(list_of_json_docs)
        except Exception as e:
            logging.error("Exception occurred in inserting json docs into mongoDB database")
            logging.exception("The error occurred is: ", e)
            
            
```
